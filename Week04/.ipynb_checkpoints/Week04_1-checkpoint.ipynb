## 【問題2】学習と検証

```python
categorical_features=[]
for col in train.columns:
    if train[col].dtype=="object":
        categorical_features.append(col)
```

```python
categorical_features
```

```python
#補間・置換(train)
null_sum = 0
for col in train.columns:
    #欠損の補間
    null_sum = train[col].isnull().sum()
    train_length = train[col].count()
    if null_sum > 0:
        if train[col].dtype == object:
            train[col] = train[col].fillna(train[col].mode()[0])
        else:
            train[col] = train[col].fillna(train[col].mean())
    
```

```python
X=train.iloc[:,2:]
y=train.iloc[:,1]
```

```python
too_many=[]
for col in categorical_features:
    if X[col].nunique()>=10:
        too_many.append(col)
print(too_many)
```

```python
X_oh=pd.get_dummies(X, columns=categorical_features)
```

```python
X_oh.shape
```

```python
y.shape
```

```python
#from sklearn.preprocessing import OneHotEncoder
#enc=OneHotEncoder(categories="auto", sparse=False, dtype=np.float32)
#oh_X=enc.fit_transform(X.values)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X_oh, y, test_size=0.25, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
Rf=RandomForestRegressor()
Rf.fit(X_train, y_train)
y_pred=Rf.predict(X_test)
```

```python
X["ORGANIZATION_TYPE"].nunique()
```

```python
from sklearn.metrics import roc_auc_score
roc=roc_auc_score(y_test, y_pred)
roc
```

```python
def fill_missing_columns(df_a, df_b):
    columns_for_b = set(df_a.columns) - set(df_b.columns)
    for column in columns_for_b:
        df_b[column] = 0
        
fill_missing_columns(X_oh, test_oh)
```
