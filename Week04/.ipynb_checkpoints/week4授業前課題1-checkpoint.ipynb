# 【課題1】練習のために特徴量とカテゴリを選択

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
#データの準備
iris=load_iris()
X=iris.data
y=iris.target.reshape(-1,1)
df=np.concatenate([X, y], axis=1)
df=pd.DataFrame(df)
col_names=iris.feature_names
col_names.append('class')
df.columns=col_names
print(df.shape)
```

```python
target_names=iris.target_names
df['class']=df['class'].apply(lambda x: target_names[int(x)])
display(df)
```

```python
df
```

```python
#virgicolorとvirginica
#sepal_lengthとpetal_length

data=df[df['class']!='setosa']
data=data[['sepal length (cm)', 'petal length (cm)', 'class']]
data
```

# 【問題3】前処理・訓練用データと検証用データの分割

```python
#目的変数を二値化
data['class']=df['class'].apply(lambda x: 0 if x=='versicolor' else 1)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split


X=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.25, random_state=1)
print(X_train.shape)
print(X_test.shape)
```

# 【問題4】前処理・標準化

```python
#標準化は訓練用のデータでfitを行い、訓練用、検証用双方のデータにtransformを行います。
#検証用データは学習の段階には存在しない未知のデータの代わりなので、標準化のための平均や標準偏差の計算に使ってはいけません。
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
std.fit(X_train)
X_train=std.transform(X_train)
X_test=std.transform(X_test)
```

```python
print(X_test)
```

# 【問題5】学習と推定

```python
from sklearn.neighbors import KNeighborsClassifier

#n_neighbors=5
KNE_5=KNeighborsClassifier(n_neighbors=5)
KNE_5.fit(X_train, y_train)
y_pred_5=KNE_5.predict(X_test)
print(y_pred_5)
print(y_test)
```

```python
#n_neighbors=1
KNE_1=KNeighborsClassifier(n_neighbors=1)
KNE_1.fit(X_train, y_train)
y_pred_1=KNE_1.predict(X_test)

#n_neighbors=3
KNE_3=KNeighborsClassifier(n_neighbors=3)
KNE_3.fit(X_train, y_train)
y_pred_3=KNE_3.predict(X_test)
```

# 【問題6】評価

https://pythondatascience.plavox.info/scikit-learn/%E5%88%86%E9%A1%9E%E7%B5%90%E6%9E%9C%E3%81%AE%E3%83%A2%E3%83%87%E3%83%AB%E8%A9%95%E4%BE%A1



### 正解率 (Accuracy)
正解率 (Accuracy) とは、「本来ポジティブに分類すべきアイテムをポジティブに分類し、本来ネガティブに分類すべきアイテムをネガティブに分類できた割合」を示し、以下の式で表されます。

Accuracy = (TP + TN) / (TP + TN + FP + FN)


### 精度 (Precision)
精度 (Precision) とは、「ポジティブに分類されたアイテムのうち、実際にポジティブであったアイテムの割合」を示し、以下の式で表されます。

Precision = TP / (TP + FP)



### 検出率 (Recall)
検出率 (Recall) とは、「本来ポジティブに分類すべきアイテムを、正しくポジティブに分類できたアイテムの割合」を示し、以下の式で表されます。
検出率は、真陽性率 (TPR, True-Positive Rate) または、感度 (Sensitivity) とも呼ばれます。

Recall = TPR = Sensitivity = TP / (TP + FN)


### F 値
F 値 (F-measure, F-score, F1 Score とも呼ばれます) とは、精度 (Precision) と検出率 (Recall) をバランス良く持ち合わせているかを示す指標です。つまり、精度は高くても、検出率が低いモデルでないか、逆に、検出率は高くても、精度が低くなっていないか、といった評価を示します。

F 値は、以下の式のように、検出精度 (Precision) と、検出率 (Recall) の調和平均で求められ、0 〜 1 の間の数値で出力され、0 の場合最も悪い評価、1 の場合最も良い評価となります。


 
F1 = 2 * (precision * recall) / (precision + recall)


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
preds=[y_pred_1, y_pred_3, y_pred_5]
n_neighbors=[1,3,5]
for pred,n in zip(preds, n_neighbors):
    accuracy=accuracy_score(y_true=y_test, y_pred=pred)
    precision=precision_score(y_test, pred)
    recall=recall_score(y_test, pred)
    f1=f1_score(y_test, pred)
    con_mat=confusion_matrix(y_test, pred)
    print('n_neightbors={}'.format(n))
    print('accuracy: ', accuracy)
    print('precision: ', precision)
    print('recall: ', recall)
    print('f1_score: ', f1)
    print('confusion matrix:\n', con_mat)
    print('-'*40)
```

# 【問題7】可視化

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches


def decision_region(X, y, model, step=0.01, title='decision region', xlabel='xlabel', ylabel='ylabel', target_names=['versicolor', 'virginica']):
    """
    2値分類を2次元の特徴量で学習したモデルの決定領域を描く。
    背景の色が学習したモデルによる推定値から描画される。
    散布図の点は訓練データまたは検証データである。

    Parameters
    ----------------
    X : ndarray, shape(n_samples, 2)
        特徴量
    y : ndarray, shape(n_samples,)
        ラベル
    model : object
        学習したモデルのインスンタスを入れる
    step : float, (default : 0.1)
        推定値を計算する間隔を設定する
    title : str
        グラフのタイトルの文章を与える
    xlabel, ylabel : str
        軸ラベルの文章を与える
    target_names= : list of str
        凡例の一覧を与える
    """
    # setting
    scatter_color = ['red', 'blue']
    contourf_color = ['pink', 'skyblue']
    n_class = 2

    # pred
    mesh_f0, mesh_f1  = np.meshgrid(np.arange(np.min(X[:,0])-0.5, np.max(X[:,0])+0.5, step), np.arange(np.min(X[:,1])-0.5, np.max(X[:,1])+0.5, step))
    mesh = np.c_[np.ravel(mesh_f0),np.ravel(mesh_f1)]
    y_pred = model.predict(mesh).reshape(mesh_f0.shape)
    

    # plot
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.contourf(mesh_f0, mesh_f1, y_pred, n_class-1, cmap=ListedColormap(contourf_color))
    plt.contour(mesh_f0, mesh_f1, y_pred, n_class-1, colors='y', linewidths=3, alpha=0.5)
    for i, target in enumerate(set(y)):
        plt.scatter(X[y==target][:, 0], X[y==target][:, 1], s=80, color=scatter_color[i], label=target_names[i], marker='o')
    patches = [mpatches.Patch(color=scatter_color[i], label=target_names[i]) for i in range(n_class)]
    plt.legend(handles=patches)
    plt.legend()
    plt.show()

```

```python
#訓練データ
decision_region(X_train, y_train, KNE_1, title='train data decision region of 1-nn', xlabel='sepal length (cm)', ylabel='petal length (cm)')
decision_region(X_train, y_train, KNE_3, title='train data decision region of 3-nn', xlabel='sepal length (cm)', ylabel='petal length (cm)')
decision_region(X_train, y_train, KNE_5, title='train data decision region of 5-nn', xlabel='sepal length (cm)', ylabel='petal length (cm)')
```

```python
#検証データ
decision_region(X_test, y_test, KNE_1, title='train data decision region of 1-nn', xlabel='sepal length (cm)', ylabel='petal length (cm)')
decision_region(X_test, y_test, KNE_3, title='train data decision region of 3-nn', xlabel='sepal length (cm)', ylabel='petal length (cm)')
decision_region(X_test, y_test, KNE_5, title='train data decision region of 5-nn', xlabel='sepal length (cm)', ylabel='petal length (cm)')
```

始めはclass名を実際の花の名前にして進めていましたが、決定境界プロットの段階で目的変数をfloatに変換する操作？があり、str型だとそれができずにエラーが出るので、問題３の冒頭に目的変数を２値化するコードを加えました。

# 【問題8】他の手法の学習

```python
#最近傍法　n_neightbors=5

y_pred=KNE_5.predict(X_test)
accuracy=accuracy_score(y_true=y_test, y_pred=y_pred)
precision=precision_score(y_test, y_pred)
recall=recall_score(y_test, y_pred)
f1=f1_score(y_test, y_pred)
con_mat=confusion_matrix(y_test, y_pred)
print('KNeighborsClassifier(n_neighbors=5)')
print('accuracy: ', accuracy)
print('precision: ', precision)
print('recall: ', recall)
print('f1_score: ', f1)
print('confusion matrix:\n', con_mat)
KNE_score=[accuracy, precision, recall, f1]
```

```python
#ロジスティック回帰
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train, y_train)
y_pred=lr.predict(X_test)
accuracy=accuracy_score(y_true=y_test, y_pred=y_pred)
precision=precision_score(y_test, y_pred)
recall=recall_score(y_test, y_pred)
f1=f1_score(y_test, y_pred)
con_mat=confusion_matrix(y_test, y_pred)
print('logistic regression')
print('accuracy: ', accuracy)
print('precision: ', precision)
print('recall: ', recall)
print('f1_score: ', f1)
print('confusion matrix:\n', con_mat)

lr_score=[accuracy, precision, recall, f1]
```

```python
#SVM
from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train, y_train)
y_pred=svc.predict(X_test)
accuracy=accuracy_score(y_true=y_test, y_pred=y_pred)
precision=precision_score(y_test, y_pred)
recall=recall_score(y_test, y_pred)
f1=f1_score(y_test, y_pred)
con_mat=confusion_matrix(y_test, y_pred)
print('SVM')
print('accuracy: ', accuracy)
print('precision: ', precision)
print('recall: ', recall)
print('f1_score: ', f1)
print('confusion matrix:\n', con_mat)

svc_score=[accuracy, precision, recall, f1]
```

```python
#決定木
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree.fit(X_train, y_train)
y_pred=tree.predict(X_test)
accuracy=accuracy_score(y_true=y_test, y_pred=y_pred)
precision=precision_score(y_test, y_pred)
recall=recall_score(y_test, y_pred)
f1=f1_score(y_test, y_pred)
con_mat=confusion_matrix(y_test, y_pred)
print('decision tree')
print('accuracy: ', accuracy)
print('precision: ', precision)
print('recall: ', recall)
print('f1_score: ', f1)
print('confusion matrix:\n', con_mat)

tree_score=[accuracy, precision, recall, f1]
```

```python
#ランダムフォレスト
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred=rf.predict(X_test)
accuracy=accuracy_score(y_true=y_test, y_pred=y_pred)
precision=precision_score(y_test, y_pred)
recall=recall_score(y_test, y_pred)
f1=f1_score(y_test, y_pred)
con_mat=confusion_matrix(y_test, y_pred)
print('random forest')
print('accuracy: ', accuracy)
print('precision: ', precision)
print('recall: ', recall)
print('f1_score: ', f1)
print('confusion matrix:\n', con_mat)

rf_score=[accuracy, precision, recall, f1]
```

```python
decision_region(X_test, y_test, KNE_5, title='train data decision region of 5-nn', xlabel='sepal length (cm)', ylabel='petal length (cm)')
decision_region(X_test, y_test, lr, title='logistic regression', xlabel='sepal length (cm)', ylabel='petal length (cm)')
decision_region(X_test, y_test, svc, title='SVM', xlabel='sepal length (cm)', ylabel='petal length (cm)')
decision_region(X_test, y_test, tree, title='decision tree', xlabel='sepal length (cm)', ylabel='petal length (cm)')
decision_region(X_test, y_test, rf, title='random forest', xlabel='sepal length (cm)', ylabel='petal length (cm)')
```

決定境界と混同行列を見ると、モデルごとの分類結果の違いは境界付近の1つのサンプルのみであることがわかる。
決定木、ランダムフォレストは分類アルゴリズムの性質上カクカクした境界になりやすいものであると思われる。
最近傍法は訓練データの分布をそのまま暗記するようなモデルなので境界に訓練データ分布の形が現れている。
評価指標値は最近傍方とSVMが最も優秀な結果となった。

```python
scores=pd.DataFrame()
scores['最近傍法']=KNE_score
scores['ロジスティック回帰']=lr_score
scores['SVM']=svc_score
scores['決定木']=tree_score
scores['ランダムフォレスト']=rf_score
scores.index=['accuracy(正解率)', 'precision(適合率)', 'recall(再現率)', 'F値']
```
