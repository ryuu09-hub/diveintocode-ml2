### **【問題4】（アドバンス課題）他の特徴量も使用しての学習**

**EDA**

```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
df.shape
```

```python
# まずはstr型データの処理
# Label Encoderを利用する

from sklearn.preprocessing import LabelEncoder

for i in range(df.shape[1]):
    if df.iloc[:, i].dtypes == object:
        lbl = LabelEncoder()
        lbl.fit(list(df.iloc[:, i].values))
        df.iloc[:, i] = lbl.transform(list(df.iloc[:, i].values))
```

```python
df.info()
```

LabelEncoderの利用により、全て数値データとなったので、続いて欠損値の処理に移る

```python
import missingno as msno
msno.matrix(df, figsize=(20, 14));
```

```python
# 欠損値が何パーセント含まれているかを確認する

total = df.isnull().sum().sort_values(ascending=False)
missing_ratio = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
missing_df = pd.concat([total, missing_ratio], axis=1, keys=['Total', 'missing ratio'])
missing_df.head(20)
```

```python
# 欠損値のある列は落とす
X_df = df.drop(['Id'], axis=1)
X = X_df.drop(['LotFrontage', 'GarageYrBlt', 'MasVnrArea'], axis=1)
X.fillna(X.median())
```

```python
corrmat = X.corr()
f, ax = plt.subplots(figsize=(20, 16))
sns.heatmap(corrmat, vmax=.8, square=True);
```

```python
# 目的変数との相関の高い特徴量を10個選び、これだけの相関係数行列ヒートマップを作る

k = 10

cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(X[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True,
                 fmt='.2f',
                 annot_kws={'size': 10},
                 yticklabels=cols.values,
                 xticklabels=cols.values)

plt.show()
```

相関の高いものはまとめて新しいfeatureを作る

```python
# GarageCarsとGarageArea
# TotalBsmtSFと1stFlrSF

X['TotalGarage'] = X['GarageCars'] + X['GarageArea']
X['TotalSF'] = X['TotalBsmtSF'] + X['1stFlrSF']
```

```python
# 目的変数の分布を確かめ、正規分布に近づける

y = df.iloc[:, -1]
ax = sns.distplot(y)
plt.show();
```

```python
# 対数変換を施す

y = np.log(y)

ax = sns.distplot(y)
plt.show();
```

大分、正規分布に近づいた

```python
# どの列が重要度の高いものかを確認する

X = X.drop(['SalePrice'], axis=1)
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=80, max_features='auto')
rf.fit(X, y)

ranking = np.argsort(-rf.feature_importances_)
f, ax = plt.subplots(figsize=(11, 9))
sns.barplot(x=rf.feature_importances_[ranking], y=X.columns.values[ranking], orient='h')
ax.set_xlabel('feature importance')
plt.tight_layout()
plt.show()
```

```python
# 目的変数との相関上位30個を抽出する
X = X.iloc[:, ranking[:30]]
```

```python
# 抽出した各特徴量と目的変数との関係をプロットする

fig = plt.figure(figsize=(12, 7))
for i in np.arange(30):
    ax = fig.add_subplot(5, 6, i+1)
    sns.regplot(x=X.iloc[:, i], y=y)

plt.tight_layout()
plt.show()
```

やはり相関がきっちりと出ているのは上位数個ということが散布図からも分かる。

**学習・推定・評価**

```python
# データの分割

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0)
```

```python
# 学習

from sklearn.svm import SVR

svr = SVR()
svr.fit(X_train, y_train)
```

```python
# テストデータを用いて目的変数を予測
y_pred_svr = svr.predict(X_test)
```

```python
# 平均２乗誤差

from sklearn.metrics import mean_squared_error

y_train_pred_svr = svr.predict(X_train)
# 学習データを用いたときの平均２乗誤差
print(mean_squared_error(y_train, y_train_pred_svr))
# テストデータを用いたときの平均２乗誤差
print(mean_squared_error(y_test, y_pred_svr))
```

```python
# 訓練データで評価
svr.score(X_train, y_train)
```

```python
# テストデータで評価
svr.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf1 = RandomForestRegressor()
rf1.fit(X_train, y_train)
```

```python
# テストデータを用いて目的変数を予測
y_pred_rf1 = rf1.predict(X_test)
```

```python
# 平均２乗誤差

from sklearn.metrics import mean_squared_error

y_train_pred_rf1 = rf1.predict(X_train)
# 学習データを用いたときの平均２乗誤差
print(mean_squared_error(y_train, y_train_pred_rf1))
# テストデータを用いたときの平均２乗誤差
print(mean_squared_error(y_test, y_pred_rf1))
```

```python
# 訓練データで評価
rf1.score(X_train, y_train)
```

```python
# テストデータで評価
rf1.score(X_test, y_test)
```

ランダムフォレストでは予測精度が非常に高くなっていることが分かる。
