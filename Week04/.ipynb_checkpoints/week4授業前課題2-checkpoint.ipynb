```python
import pandas as pd
df=pd.read_csv('/Users/wakaichiaki/diveintocode-ml/diveintocode-ml/Week04/train.csv')
```

```python
print(df.shape)
X=df[['GrLivArea', 'YearBuilt']]
y=df['SalePrice']
print(X.head())
print(y.head())
```

# 【問題2】線形回帰による推定と評価

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.25, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
linreg.fit(X_train, y_train)
y_pred=linreg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
lin_mse=mean_squared_error(y_test, y_pred)
print('平均２乗誤差: ', lin_mse)
```

```python
fig, ax=plt.subplots(1,2,figsize=(15,5))
ax[0].scatter(X_test['GrLivArea'], y_test, color='orange', label='true')
ax[0].scatter(X_test['GrLivArea'], y_pred, color='blue', label='prediction')
ax[0].set_xlabel('GrLivArea')
ax[0].set_ylabel('Sale Price')
ax[0].legend(loc='upper left')

ax[1].scatter(X_test['YearBuilt'], y_test, color='orange', label='true')
ax[1].scatter(X_test['YearBuilt'], y_pred, color='blue', label='prediction')
ax[1].set_xlabel('Year Build')
ax[1].set_ylabel('Sale Price')
ax[1].legend(loc='upper left')
plt.show()
```

# 【問題3】手法の比較

https://pythondatascience.plavox.info/scikit-learn/%E5%9B%9E%E5%B8%B0%E3%83%A2%E3%83%87%E3%83%AB%E3%81%AE%E8%A9%95%E4%BE%A1
平均２乗誤差以外の評価指標については上記のサイトを参考に平均絶対誤差(MAE), 二乗平均平方根誤差 (RMSE), 決定係数 (R2)を利用する。

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

lin_mae=mean_absolute_error(y_test, y_pred)
lin_rmse=mean_squared_error(y_test, y_pred)
lin_r2=r2_score(y_test, y_pred)

metrics_names=['平均二乗誤差(MSE)', '平均絶対誤差(MAE)', '二乗平均平方根誤差(RMSE)', '決定係数(R2)']
lin_metrics=[lin_mse, lin_mae, lin_rmse, lin_r2]
```

### SVM

```python
from sklearn.svm import SVR
svr=SVR(kernel='linear')
svr.fit(X_train, y_train)
y_pred_s=svr.predict(X_test)
s_mse=mean_squared_error(y_test, y_pred_s)
s_mae=mean_absolute_error(y_test, y_pred_s)
s_rmse=mean_squared_error(y_test, y_pred_s)
s_r2=r2_score(y_test, y_pred_s)
s_metrics=[s_mse, s_mae, s_rmse, s_r2]
print(s_metrics)
```

```python
fig, ax=plt.subplots(1,2,figsize=(15,5))
ax[0].scatter(X_test['GrLivArea'], y_test, color='orange', label='true')
ax[0].scatter(X_test['GrLivArea'], y_pred_s, color='blue', label='prediction')
ax[0].set_xlabel('GrLivArea')
ax[0].set_ylabel('Sale Price')
ax[0].legend(loc='upper left')

ax[1].scatter(X_test['YearBuilt'], y_test, color='orange', label='true')
ax[1].scatter(X_test['YearBuilt'], y_pred_s, color='blue', label='prediction')
ax[1].set_xlabel('Year Build')
ax[1].set_ylabel('Sale Price')
ax[1].legend(loc='upper left')
plt.show()
```

### 決定木

```python
from sklearn.tree import DecisionTreeRegressor
tree=DecisionTreeRegressor()
tree.fit(X_train, y_train)
y_pred_t=tree.predict(X_test)
t_mse=mean_squared_error(y_test, y_pred_t)
t_mae=mean_absolute_error(y_test, y_pred_t)
t_rmse=mean_squared_error(y_test, y_pred_t)
t_r2=r2_score(y_test, y_pred_t)
t_metrics=[t_mse, t_mae, t_rmse, t_r2]
```

```python
fig, ax=plt.subplots(1,2,figsize=(15,5))
ax[0].scatter(X_test['GrLivArea'], y_test, color='orange', label='true')
ax[0].scatter(X_test['GrLivArea'], y_pred_t, color='blue', label='prediction')
ax[0].set_xlabel('GrLivArea')
ax[0].set_ylabel('Sale Price')
ax[0].legend(loc='upper left')

ax[1].scatter(X_test['YearBuilt'], y_test, color='orange', label='true')
ax[1].scatter(X_test['YearBuilt'], y_pred_t, color='blue', label='prediction')
ax[1].set_xlabel('Year Build')
ax[1].set_ylabel('Sale Price')
ax[1].legend(loc='upper left')
plt.show()
```

### ランダムフォレスト

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(X_train, y_train)
y_pred_r=rf.predict(X_test)
r_mse=mean_squared_error(y_test, y_pred_r)
r_mae=mean_absolute_error(y_test, y_pred_r)
r_rmse=mean_squared_error(y_test, y_pred_r)
r_r2=r2_score(y_test, y_pred_r)
r_metrics=[r_mse, r_mae, r_rmse, r_r2]
```

```python
fig, ax=plt.subplots(1,2,figsize=(15,5))
ax[0].scatter(X_test['GrLivArea'], y_test, color='orange', label='true')
ax[0].scatter(X_test['GrLivArea'], y_pred_r, color='blue', label='prediction')
ax[0].set_xlabel('GrLivArea')
ax[0].set_ylabel('Sale Price')
ax[0].legend(loc='upper left')

ax[1].scatter(X_test['YearBuilt'], y_test, color='orange', label='true')
ax[1].scatter(X_test['YearBuilt'], y_pred_r, color='blue', label='prediction')
ax[1].set_xlabel('Year Build')
ax[1].set_ylabel('Sale Price')
ax[1].legend(loc='upper left')
plt.show()
```

```python
metrics_df=pd.DataFrame()
metrics_df['線形回帰']=lin_metrics
metrics_df['SVM']=s_metrics
metrics_df['決定木']=t_metrics
metrics_df['ランダムフォレスト']=r_metrics
metrics_df.index=metrics_names
```

```python
metrics_df
```

グラフおよび表を見ると、線形回帰とSVM、決定木とランダムフォレストがそれぞれ類似した結果になったことがわかる。
MSE,MAE,RMSEは前者のモデルの方が小さかったが、R2のみ後者のモデルの方が小さい。
R2は0から１の値を取り、予測値が観測値に近いほど１に近くと言った評価指標であるため、今回は全ての評価指標において線形回帰、SVMが良い結果になったと言える。
グラフから、線形回帰とSVMは直線に沿ったようなプロットになっているが、一方決定木とランダムフォレストは直線上から外れたデータにも対応しようとしていることがわかる(グラフ右上など)。
