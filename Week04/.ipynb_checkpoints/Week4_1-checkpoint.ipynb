**4種類の指標値（Accuracy, Precision, Recall, F値）の意味**

・真陽性（TP: True Positive）：実際のクラスが陽性で予測も陽性（正解）

・真陰性（TN: True Negative）：実際のクラスが陰性で予測も陰性（正解）

・偽陽性（FP: False Positive）：実際のクラスは陰性で予測が陽性（不正解）

・偽陰性（FN: False Negative）：実際のクラスは陽性で予測が陰性（不正解）

**Accuracy（正解率）**

　正解率とは、全てのサンプルのうち正解したサンプルの割合。式にすると以下。
 
 $$ Accuracy = \frac{TP + TN}{TP + TN + FP + FN} $$
 
**Precision(適合率)**
 
 　適合率とは、陽性と予測されたサンプルのうち正解したサンプルの割合。FP(偽陽性)と適合率は反比例する。偽陽性を避けたい場合に用いる指標。式は以下。
  
 $$ Precision = \frac{TP}{TP + FP} $$
  
**Recall(再現率)**

　再現率とは、実際に陽性のサンプルのうち正解したサンプルの割合。FN(偽陰性)と再現率は反比例する。偽陰性を避けたい場合に用いる指標。式は以下。
 
 $$ Recall = \frac{TP}{TP + FN} $$
 
**F1値(F1-measure)**

　F1値とは、適合率と再現率の調和平均。式は以下。
 
 $$ F1-measure = \frac{2 * Precision * Recall}{Precision + Recall} = \frac{2 * TP}{2 * TP + FP + FN} $$

### **【問題7】可視化**

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches

def decision_region(X, y, model, step=0.01, title='decision region', xlabel='xlabel', ylabel='ylabel', target_names=['versicolor', 'virginica']):
    """
    2値分類を2次元の特徴量で学習したモデルの決定領域を描く。
    背景の色が学習したモデルによる推定値から描画される。
    散布図の点は訓練データまたは検証データである。

    Parameters
    ----------------
    X : ndarray, shape(n_samples, 2)
        特徴量
    y : ndarray, shape(n_samples,)
        ラベル
    model : object
        学習したモデルのインスンタスを入れる
    step : float, (default : 0.1)
        推定値を計算する間隔を設定する
    title : str
        グラフのタイトルの文章を与える
    xlabel, ylabel : str
        軸ラベルの文章を与える
    target_names= : list of str
        凡例の一覧を与える
    """
    # setting
    scatter_color = ['red', 'blue']
    contourf_color = ['pink', 'skyblue']
    n_class = 2

    # pred
    mesh_f0, mesh_f1  = np.meshgrid(np.arange(np.min(X[:,0])-0.5, np.max(X[:,0])+0.5, step), np.arange(np.min(X[:,1])-0.5, np.max(X[:,1])+0.5, step))
    mesh = np.c_[np.ravel(mesh_f0),np.ravel(mesh_f1)]
    y_pred = model.predict(mesh).reshape(mesh_f0.shape)

    # plot
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.contourf(mesh_f0, mesh_f1, y_pred, n_class-1, cmap=ListedColormap(contourf_color))
    plt.contour(mesh_f0, mesh_f1, y_pred, n_class-1, colors='y', linewidths=3, alpha=0.5)
    for i, target in enumerate(set(y)):
        plt.scatter(X[y==target][:, 0], X[y==target][:, 1], s=80, color=scatter_color[i], label=target_names[i], marker='o')
    patches = [mpatches.Patch(color=scatter_color[i], label=target_names[i]) for i in range(n_class)]
    plt.legend(handles=patches)
    plt.legend()
    plt.show()
```

```python
decision_region(X_std, y, knc, step=0.01, title='decision region of 5-nn', xlabel='sepal length', ylabel='petal length', target_names=['versicolor', 'virginica'])
```

```python
# 検証用データをプロット

decision_region(X_test, y_test, knc, step=0.01, title='decision region of 5-nn', xlabel='sepal length', ylabel='petal length', target_names=['versicolor', 'virginica'])
```

```python
# 可視化（多値分類版）

def decision_region3(X, y, model, step=0.01, title='decision region', xlabel='xlabel', ylabel='ylabel', target_names=['versicolor', 'virginica']):
    """
    2値分類を2次元の特徴量で学習したモデルの決定領域を描く。
    背景の色が学習したモデルによる推定値から描画される。
    散布図の点は訓練データまたは検証データである。

    Parameters
    ----------------
    X : ndarray, shape(n_samples, 2)
        特徴量
    y : ndarray, shape(n_samples,)
        ラベル
    model : object
        学習したモデルのインスンタスを入れる
    step : float, (default : 0.1)
        推定値を計算する間隔を設定する
    title : str
        グラフのタイトルの文章を与える
    xlabel, ylabel : str
        軸ラベルの文章を与える
    target_names= : list of str
        凡例の一覧を与える
    """
    # setting
    scatter_color = ['green', 'red', 'blue']
    contourf_color = ['lightgreen', 'pink', 'skyblue']
    n_class = 3

    # pred
    mesh_f0, mesh_f1  = np.meshgrid(np.arange(np.min(X[:,0])-0.5, np.max(X[:,0])+0.5, step), np.arange(np.min(X[:,1])-0.5, np.max(X[:,1])+0.5, step))
    mesh = np.c_[np.ravel(mesh_f0),np.ravel(mesh_f1)]
    y_pred = model.predict(mesh).reshape(mesh_f0.shape)

    # plot
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.contourf(mesh_f0, mesh_f1, y_pred, n_class-1, cmap=ListedColormap(contourf_color))
    plt.contour(mesh_f0, mesh_f1, y_pred, n_class-1, colors='y', linewidths=3, alpha=0.5)
    for i, target in enumerate(set(y)):
        plt.scatter(X[y==target][:, 0], X[y==target][:, 1], s=80, color=scatter_color[i], label=target_names[i], marker='o')
    patches = [mpatches.Patch(color=scatter_color[i], label=target_names[i]) for i in range(n_class)]
    plt.legend(handles=patches)
    plt.legend()
    plt.show()
```

```python
decision_region3(X3_train, y3_train, knc3, step=0.01, title='decision region', xlabel='sepal length', ylabel='petal length', target_names=['setosa', 'versicolor', 'virginica'])
```

```python
decision_region3(X3_test, y3_test, knc3, step=0.01, title='decision region', xlabel='sepal length', ylabel='petal length', target_names=['setosa', 'versicolor', 'virginica'])
```

どのクラスも高い精度で分類ができているが、クラス0（setosa）の分類は100%正しく分類できていることが分かる。また、上の決定領域の可視化からもその妥当性が確認できる。

### 【まとめ】

以上のことからSVMが最も精度の高い手法という結果になった。各手法にはそれぞれ得意とする問題がありそうなことが伺え、例えば線形分離可能な問題ではロジスティック回帰がその威力を発揮するが線形分離不可能な問題では途端にその精度は下がる。SVMはどのような問題にも高い精度で対応することができる印象を受けた。近く最適化数学の知識をより深め、SVMの理論を理解したい。
