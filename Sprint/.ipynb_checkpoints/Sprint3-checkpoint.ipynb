## 線形回帰スクラッチ


```python
import numpy as np
import pandas as pd 

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from numpy.random import *
```

```python
class ScratchLinearRegression():
    """
    線形回帰のスクラッチ実装

    Parameters
    ----------
    num_iter : int
      イテレーション数
    lr : float
      学習率
    no_bias : bool
      バイアス項を入れない場合はTrue
    verbose : bool
      学習過程を出力する場合はTrue

    Attributes
    ----------
    self.coef_ : 次の形のndarray, shape (n_features,)
      パラメータ
    self.loss : 次の形のndarray, shape (self.iter,)
      学習用データに対する損失の記録
    self.val_loss : 次の形のndarray, shape (self.iter,)
      検証用データに対する損失の記録

    """

    def __init__(self, num_iter, lr, no_bias, verbose):
        # ハイパーパラメータを属性として記録
        self.iter = num_iter
        self.lr = lr
        self.no_bias = no_bias
        self.verbose = verbose
        # 損失を記録する配列を用意
        self.loss = np.zeros(self.iter)
        self.val_loss = np.zeros(self.iter)

    def fit(self, X, y, X_val=None, y_val=None):
        """
        線形回帰を学習する。検証用データが入力された場合はそれに対する損失と精度もイテレーションごとに計算する。

        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            学習用データの特徴量
        y : 次の形のndarray, shape (n_samples, )
            学習用データの正解値
        X_val : 次の形のndarray, shape (n_samples, n_features)
            検証用データの特徴量
        y_val : 次の形のndarray, shape (n_samples, )
            検証用データの正解値
        """
        #バイアスの有無に対する場合分け
        if self.no_bias == True:
            self.bias=0 
        else: 
            self.bias=np.random.rand()
        
        #thetaの定義
        self.theta=np.random.randn(X.shape[1])
        
        #サンプル数の定義
        self.m=X.shape[0]
        
        #学習をself.iter回 繰り返す
        for i in range(self.iter):
            #最急降下法によるシータとバイアスの更新
            self._gradient_descent(X, y)
            loss1=((self._linear_hypothesis(X)-y)**2).sum()
            
            
            
            #【問題5】目的関数 
            cost_function= 0.5 * self.m*loss1
            
            #平均二乗誤差で評価し損失を記録
            self.loss[i]=self.MSE(X, y)
            self.val_loss[i]=self.MSE(X_val, y_val)
            if self.verbose==True:
            #verboseをTrueにした際は学習過程を出力
                    print("{}回目:MSE {}".format(i+1,self.loss[i]))


    def predict(self, X):
        """
        線形回帰を使い推定する。

        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            サンプル

        Returns
        -------
            次の形のndarray, shape (n_samples, 1)
            線形回帰による推定結果
        """
        
        return self._linear_hypothesis(X)

    #【問題1】仮定関数
    def _linear_hypothesis(self, X):
        """
        線形の仮定関数を計算する

        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
          学習データ

        Returns
        -------
          次の形のndarray, shape (n_samples, 1)
          線形の仮定関数による推定結果

        """
    
        h=np.dot(self.theta, X.T) + self.bias
        
        return h
    
    #【問題4】平均二乗誤差
    def MSE(self, X, y):
        """
        平均二乗誤差の計算

        Parameters
        ----------
        y_pred : 次の形のndarray, shape (n_samples,)
          推定した値
        y : 次の形のndarray, shape (n_samples,)
          正解値

        Returns
        ----------
        mse : numpy.float
          平均二乗誤差
        """

        loss2=((self._linear_hypothesis(X)-y)**2).sum()
        mse=1/X.shape[0]*loss2
    
        return mse
    
    #【問題2】最急降下法
    def _gradient_descent(self, X, y):
        """
        説明を記述
        """
        
        self.theta=self.theta - (self.lr /self.m) * ((self._linear_hypothesis(X)-y).reshape(-1, 1) * X).sum(axis=0)
        
        if not self.no_bias:
            self.bias=self.bias -(self.lr /self.m) * (self._linear_hypothesis(X)-y).reshape(-1, 1)
        
    
```

## データセットロード

```python
df=pd.read_csv('/Users/wakaichiaki/diveintocode-ml/diveintocode-ml/Week04/train.csv')
```

```python
X=df[['GrLivArea']].values
y=df['SalePrice'].values
```

```python
plt.title("GrLivArea-SalePrice")
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.scatter(X, y)
plt.show()
```

```python
model=ScratchLinearRegression(2000, 0.000000001, True, True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.25,random_state=0)
```

```python
#学習させてみる
model.fit(X_train, y_train, X_test, y_test)
```

```python
model.predict(X_test)
```

## 用語確認

【前提知識】

ディープラーニングのように内部で確率的勾配降下法を使う場合、データセットを幾つかのサブセットに分けて学習する。(学習する際に異常値の影響を小さくするため)

**ハイパーパラメータ**
学習前に設定する(重みやバイアスとは区別)
- 学習率　
- 重み初期値（±1, Xavier, Gaussian, He）
- 活性化関数（Sigmoid，Tanh, ReLU）
- 一括重み更新、逐次的重み更新、ミニバッチ処理 
- レイヤー数、ニューロン数 
- パラメータの更新方法（SGD、Adam、AdGard、Momentam、SMPROS、RMSProp）
- 正規化

**イテレーション数**

データセットに含有データが少なくとも１回は学習に用いられるのに必要な学習回数
→イテレーション数=データサイズ/バッチサイズ（イテレーション数=サブセット数）

**エポック数**

データセットの含有データを少なくとも1回は学習に用いる手順を1エポックと呼ぶ。
エポック数は、損失関数（コスト関数）の値がほぼ収束するまで行うのが良い。(※完全収束まで行うと過学習になる恐れがある)


**学習率**
重みのパラメータを学習していく中で更新していく

**損失関数**
平均二乗誤差 / Mean Squared Errorのこと
